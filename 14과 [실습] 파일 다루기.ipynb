```python
a = open("poem.txt", 'r')

print(a.readlines())
a.close()
```

```python
file = open('waterloo.txt', "r", encoding='UTF8')
text = file.read() 
file.close()

wordList = text.split()
```

```python
print(wordList)
```

```python
worduniq = set(wordList)

wordCount = {}

for w in worduniq:
    wordCount[w] = wordList.count(w)

del_word = ['the','a','is','of','on','that','this','and','be','to','in','have','from','not','as','by','which','into','words','it',]
for w in del_word :
    if w in wordCount :
        del wordCount[w]
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(width = 1000, height = 600).generate_from_frequencies(wordCount)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.show()
```
